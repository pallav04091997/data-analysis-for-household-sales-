```python
'''
1 - to count overlapping urlh
'''
import pandas as pd  #just for reading the data
t = pd.read_csv('texcel.csv') 
y= pd.read_csv('yexcel.csv', encoding = "ISO-8859-1")
t_urlh=[] #list to store urlh for t.csv
for i in t.urlh:
    t_urlh.append(i)
y_urlh = []
for i in y.urlh:
    y_urlh.append(i)
overlap_count=0
for i in range(len(t_urlh)):
    if t_urlh[i] ==y_urlh[i]:
        overlap_count+=1
        
    else:
        pass
print(overlap_count)
```

```python
'''
2 - For all the URLH which are overlapping, calculate the price difference
'''
unique_urlh_t = []
unique_urlh_y = []
index_t = []
index_y = []

for i,j in enumerate(t.urlh):
    if j in unique_urlh_t:
        pass
    else:
        unique_urlh_t.append(j)
        index_t.append(i)
        
for i,j in enumerate(y.urlh):
    if j in unique_urlh_y:
        pass
    else:
        unique_urlh_y.append(j)
        index_y.append(i)
diff_price= []
updated_index = []
for i in range(0,606):
    if index_t[i]==index_y[i]:
        if unique_urlh_t[i]==unique_urlh_y[i]:
            diff_price.append(abs(t['available_price'][i] - y['available_price'][i]))
            updated_index.append(i)
            
print('index    urlh')
for i in range(len(updated_index)):
    print(f"{updated_index[i]} \t {diff_price[i]} ")  
```

```python
'''
3 - no. of unique category
alternate method could be using pandas pre-defined nunique() function which would have given the answer in one line.
'''
unique_category = []
for i in t.category:
    if i in unique_category:
        pass
    else:
        unique_category.append(i)
for i in y.category:
    if i in unique_category:
        pass
    else:
        unique_category.append(i)
print(f"number of unique categories in t and y are {len(unique_category)}")
print(f"unique categories in file t and y are {unique_category}")
```

```python
'''
4- List of categories which is not overlapping.
'''
not_overlapping_category = []
t_cat = []
y_cat = []
for i in t.category:
    t_cat.append(i)
for i in y.category:
    y_cat.append(i)
for i in range(len(t_cat)):
    if t_cat[i] == y_cat[i]:
        pass
    else:
        not_overlapping_category.append(t_cat[i])
        not_overlapping_category.append(y_cat[i])
        print(f"different category at index {i} and category for file t is {t_cat[i]} and for file y is {y_cat[i]}")

print('\n\n\n',not_overlapping_category)
```

```python
'''
5- Generate the stats with count for all taxonomies
'''
category_subcategory = []
for i,j in enumerate(t.category):
    category_subcategory.append(j+'>'+t['subcategory'][i])
for i,j in enumerate(y.category):
    category_subcategory.append(j+'>'+y['subcategory'][i])
    
set_1 = set(category_subcategory)
dict_1 = {}
for i in set_1:
    dict_1[i] = 0

for i in category_subcategory:
    dict_1[i] = dict_1.get(i) + 1
    
for i,j in dict_1.items():
    print(i, ':', j)

import matplotlib.pyplot as plt
plt.bar(range(len(dict_1)), list(dict_1.values()))
plt.xticks(range(len(dict_1)), list(dict_1.keys()), rotation = 90)


plt.show()
```

```python
'''
6 - normalize mrp
'''
normalize_list_t = []
for i,j in enumerate(t.mrp):
    if j == 0 or j is None:
        normalize_list_t.append("NA")
    else :
        val = 1-(t['mrp'][i]/max(t.mrp))
        normalize_list_t.append(val)
        
with open('normalize_mrp_t.csv', 'w') as f:
    for j in normalize_list_t:
        f.write('%s\n' % j)
        
        
######################################################

normalize_list_y = []
for i,j in enumerate(y.mrp):
    if j == 0 or j is None:
        normalize_list_y.append("NA")
    else :
        value = 1-(y['mrp'][i]/max(y.mrp))
        normalize_list_y.append(value)
        
with open('normalize_mrp_y.csv', 'w') as f:
    for j in normalize_list_y:
        f.write('%s\t' % j)
       
    
```
